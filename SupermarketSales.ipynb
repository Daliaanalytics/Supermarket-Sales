```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data= {
    'Invoice_Id':['INV-001','INV-002','INV-003','INV-004','INV-005',
                   'INV-006','INV-007','INV-008','INV-009','INV-010',
                   'INV-011','INV-012','INV-013','INV-014','INV-015'],
    'Branch': ['A','B','C','A','B','C','A','B','C','A','B','C','A','B','C'],
    'City':['Cairo','Alex','Giza','Cairo','Alex','Giza','Cairo','Alex','Giza','Cairo','Alex','Giza','Cairo','Alex','Giza'],
    'Category':['Electronics','Food','Fashion','Electronics','Fashion','Food',
                 'Food','Electronics','Fashion','Fashion','Food','Electronics',
                 'Electronics','Fashion','Food'],
    'Quantity': [3,2,5,1,4,2,3,1,6,2,3,4,2,5,1],
    'Unit_Price': [500,50,200,1500,300,80,60,800,150,250,70,600,900,180,90],
    'Payment':['Cash','Credit','Cash','Ewallet','Cash','Credit',
                'Cash','Ewallet','Credit','Cash','Credit','Ewallet',
                'Cash','Credit','Cash'],
    'Rating':[8.5,7.2,9.0,6.5,8.8,7.5,9.2,6.8,8.1,7.9,8.3,9.1,7.6,8.7,6.9],
    'Month': ['Jan','Jan','Jan','Feb','Feb','Feb','Mar','Mar','Mar',
              'Apr','Apr','Apr','May','May','May']
}
df= pd.DataFrame(data)
df['Total']= df['Quantity']* df['Unit_Price']
```

```python
City_Sales= df.groupby('City')['Total'].sum().sort_values(ascending=False)
plt.bar(City_Sales.index, City_Sales.values)
plt.title('Total Sales by City')
plt.xlabel('City')
plt.ylabel('Total')
plt.savefig('Save_by_city.png')
plt.show()
```
